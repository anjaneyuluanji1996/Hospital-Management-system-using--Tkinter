```python
import tkinter as tk
from tkinter import messagebox
import mysql.connector
from mysql.connector import Error
from datetime import datetime
```

```python
## Project Title : Hospital Management System using Tkinter

def connect_db(db=None):
    try:
        return mysql.connector.connect(
            host="localhost",
            user="root",
            password="Anjaneyulu@4726",
            database="hospital_db"
        )
    except Error as e:
        messagebox.showerror("Database Error", str(e))
        return None

def setup_database():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS hospital_db")
    conn.close()

    conn = connect_db("hospital_db")
    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS patients (
        patient_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(100),
        age INT,
        gender VARCHAR(10),
        contact VARCHAR(15),
        disease VARCHAR(100),
        admit_date DATE
    )""")

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS doctors (
        doctor_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(100),
        specialization VARCHAR(100),
        contact VARCHAR(15)
    )""")

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS appointments (
        appointment_id INT AUTO_INCREMENT PRIMARY KEY,
        patient_id INT,
        doctor_id INT,
        appointment_date DATE
    )""")

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS pharmacy (
        medicine_id INT AUTO_INCREMENT PRIMARY KEY,
        medicine_name VARCHAR(100),
        quantity INT,
        price DECIMAL(10,2)
    )""")

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS billing (
        bill_id INT AUTO_INCREMENT PRIMARY KEY,
        patient_id INT,
        amount DECIMAL(10,2),
        bill_date DATE
    )""")

    conn.commit()
    conn.close()


def patient_window():
    win = tk.Toplevel(root)
    win.title("Add Patient")

    fields = ["Name", "Age", "Gender", "Contact", "Disease"]
    entries = []

    for i, f in enumerate(fields):
        tk.Label(win, text=f).grid(row=i, column=0)
        e = tk.Entry(win)
        e.grid(row=i, column=1)
        entries.append(e)

    def save_patient():
        conn = connect_db("hospital_db")
        cursor = conn.cursor()
        cursor.execute("""
            INSERT INTO patients(name,age,gender,contact,disease,admit_date)
            VALUES(%s,%s,%s,%s,%s,%s)
        """, (
            entries[0].get(),
            entries[1].get(),
            entries[2].get(),
            entries[3].get(),
            entries[4].get(),
            datetime.now().date()
        ))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Patient Added")

    tk.Button(win, text="Save Patient", command=save_patient).grid(row=6, column=1)


def doctor_window():
    win = tk.Toplevel(root)
    win.title("Add Doctor")

    fields = ["Name", "Specialization", "Contact"]
    entries = []

    for i, f in enumerate(fields):
        tk.Label(win, text=f).grid(row=i, column=0)
        e = tk.Entry(win)
        e.grid(row=i, column=1)
        entries.append(e)

    def save_doctor():
        conn = connect_db("hospital_db")
        cursor = conn.cursor()
        cursor.execute("""
            INSERT INTO doctors(name,specialization,contact)
            VALUES(%s,%s,%s)
        """, (
            entries[0].get(),
            entries[1].get(),
            entries[2].get()
        ))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Doctor Added")

    tk.Button(win, text="Save Doctor", command=save_doctor).grid(row=4, column=1)


def appointment_window():
    win = tk.Toplevel(root)
    win.title("Book Appointment")

    labels = ["Patient ID", "Doctor ID", "Date (YYYY-MM-DD)"]
    entries = []

    for i, l in enumerate(labels):
        tk.Label(win, text=l).grid(row=i, column=0)
        e = tk.Entry(win)
        e.grid(row=i, column=1)
        entries.append(e)

    def book():
        conn = connect_db("hospital_db")
        cursor = conn.cursor()
        cursor.execute("""
            INSERT INTO appointments(patient_id,doctor_id,appointment_date)
            VALUES(%s,%s,%s)
        """, (
            entries[0].get(),
            entries[1].get(),
            entries[2].get()
        ))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Appointment Booked")

    tk.Button(win, text="Book Appointment", command=book).grid(row=4, column=1)


def pharmacy_window():
    win = tk.Toplevel(root)
    win.title("Add Medicine")

    labels = ["Medicine Name", "Quantity", "Price"]
    entries = []

    for i, l in enumerate(labels):
        tk.Label(win, text=l).grid(row=i, column=0)
        e = tk.Entry(win)
        e.grid(row=i, column=1)
        entries.append(e)

    def save_medicine():
        conn = connect_db("hospital_db")
        cursor = conn.cursor()
        cursor.execute("""
            INSERT INTO pharmacy(medicine_name,quantity,price)
            VALUES(%s,%s,%s)
        """, (
            entries[0].get(),
            entries[1].get(),
            entries[2].get()
        ))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Medicine Added")

    tk.Button(win, text="Save Medicine", command=save_medicine).grid(row=4, column=1)


def billing_window():
    win = tk.Toplevel(root)
    win.title("Generate Bill")

    labels = ["Patient ID", "Amount"]
    entries = []

    for i, l in enumerate(labels):
        tk.Label(win, text=l).grid(row=i, column=0)
        e = tk.Entry(win)
        e.grid(row=i, column=1)
        entries.append(e)

    def generate_bill():
        conn = connect_db("hospital_db")
        cursor = conn.cursor()
        cursor.execute("""
            INSERT INTO billing(patient_id,amount,bill_date)
            VALUES(%s,%s,%s)
        """, (
            entries[0].get(),
            entries[1].get(),
            datetime.now().date()
        ))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Bill Generated")

    tk.Button(win, text="Generate Bill", command=generate_bill).grid(row=3, column=1)


root = tk.Tk()
root.title("Hospital Management System")
root.geometry("400x450")

tk.Label(root, text="HOSPITAL MANAGEMENT SYSTEM",
         font=("Arial", 16, "bold")).pack(pady=20)

tk.Button(root, text="Add Patient", width=30, command=patient_window).pack(pady=5)
tk.Button(root, text="Add Doctor", width=30, command=doctor_window).pack(pady=5)
tk.Button(root, text="Book Appointment", width=30, command=appointment_window).pack(pady=5)
tk.Button(root, text="Add Medicine", width=30, command=pharmacy_window).pack(pady=5)
tk.Button(root, text="Generate Bill", width=30, command=billing_window).pack(pady=5)
tk.Button(root, text="Exit", width=30, command=root.quit).pack(pady=15)

# setup DB AFTER window loads
root.after(100, setup_database)

root.mainloop()
```
